```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
plt.style.use('default')
from scipy.signal import argrelextrema
```

# 1. Lamps spectrums

## 1.1 Argon

```python
df = pd.read_csv('Lamps spectrums/Ar.esp', delimiter=' ', names=('Wavelength', 'Value'), skiprows=[0, 1])
df.head()
```

```python
n = 1 # Number of points to be checked before and after.
# Find local peaks
df['max'] = df.iloc[argrelextrema(df['Value'].values, np.greater_equal, order=n)[0]]['Value']
df_assist = df[df['max'] > 1000] # Filter values.
df_assist
```

```python
plt.figure(figsize=(12, 8))
plt.title('Argon spectrum')
plt.plot(df['Wavelength'], df['Value'])
plt.plot(df_assist['Wavelength'], df_assist['Value'], 'o')
plt.xlabel('Wavelenght, nm')
plt.ylabel('Intensivity, counts per frame')
plt.grid()
plt.show()
plt.savefig('..\Spectrometer/Pictures/argon.pdf')
```

# 3 'Sun' spectrums

```python
df_1 = pd.read_csv('\'Sun\' spectrums/\'Sun\'.esp', delimiter=' ', names=('Wavelength', 'Value'), skiprows=[0, 1])
df_2 = pd.read_csv('\'Sun\' spectrums/\'Sun\' with glass.esp', delimiter=' ', names=('Wavelength', 'Value'), skiprows=[0, 1])
```

```python
plt.figure(figsize=(12, 8))
plt.title('Gray sky spectrum')
plt.plot(df_1['Wavelength'], df_1['Value'], alpha=0.5, label='No glass')
plt.plot(df_2['Wavelength'], df_2['Value'], alpha=0.5, label='With glass')
plt.xlabel('Wavelenght, nm')
plt.ylabel('Intensivity, counts per frame')
plt.grid()
plt.legend()
plt.show()
plt.savefig('..\Spectrometer/Pictures/sky_spectrum.pdf')
```

```python
plt.figure(figsize=(12, 8))
plt.title('Dependance of glass permeability on wavelength')
plt.plot(df_1['Wavelength'], (df_2['Value']/df_1['Value']))
plt.xlabel('Wavelenght, nm')
plt.ylabel('Intensivity, counts per frame')
plt.grid()
plt.show()
plt.savefig('..\Spectrometer/Pictures/glass_permeability.pdf')
```
