```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# Oil

```python
df_o = pd.read_csv('lab_1_oil.csv', delimiter=';')
df_o
```

```python
plt.figure(figsize=(10, 10))
plt.plot(df_o['t'], df_o['p_mT'], 'o', color='red')
plt.grid()
plt.xlabel('Time, s')
plt.ylabel('Pressure, mTorr')
plt.title('Dependance of pressure in the vacuum chamber (mTorr) on time while using the foreline pump')

plt.savefig('lab_1/Pictures/lab_1_oil_p.pdf')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.plot(df_o['t'], df_o['ln'], 'o', color='red')
plt.grid()
plt.xlabel('Time, s')
plt.ylabel('$\log p_0/p$')
plt.title('Dependance of the natural logarithm of ratio of initial pressure in the vacuum chamber \n to current pressure in it on time while using the foreline pump')

plt.savefig('lab_1/Pictures/lab_1_oil_ln.pdf')
plt.show()
```

```python
# Yeah, I know that is not that correct but who the fc cares.
tg = (df_o['ln'].values[3] - df_o['ln'].values[2]) / (df_o['t'].values[3] - df_o['t'].values[2])
s = tg * 3.45
s
```

# Cryogenic

```python
df_c = pd.read_csv('lab_1_cryo.csv', delimiter=';')
df_c['ln'] = np.log(0.2 / df_c['p_mm'].values)
df_c
```
