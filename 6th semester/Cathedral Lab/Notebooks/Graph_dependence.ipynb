```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit

import matplotlib
font = {'weight' : 'normal',
        'size'   : 17}
matplotlib.rc('font', **font)
```

```python
def linear(x, a, b):
    return a * x + b

def exponent(x, a, b):
    return np.exp(-x * a) + b
```

```python
voltages = [25.65, 64.4, 93.3, 150.1, 200]
values = [16.7156862745, 23.2352941176, 38.2352941176, 46.1, 59.9509803922]
values.reverse()
voltages = np.array(voltages)
values = np.array(values)
```

```python
parameters = curve_fit(linear, voltages, np.log(values))[0]
x_assist = np.linspace(voltages[0] * 0.9, voltages[-1] * 1.1, int(1e3))
```

```python
plt.figure(figsize=(10, 10))
plt.plot(voltages, np.log(values), 'o', label='Эксперимент')
plt.legend()
plt.plot(x_assist, linear(x_assist, *parameters), '-')

plt.xlabel('Напряжение $U$, В')
plt.ylabel('Логарифм глубины профиля $\ln Z$')

plt.grid()

plt.savefig('../Report/Pictures/Curve_profiles.pdf', bbox_inches='tight')
plt.savefig('../Report/Pictures/Curve_profiles.png', bbox_inches='tight')
plt.show()
```
